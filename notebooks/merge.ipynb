```python
import pandas as pd
```

```python
all_countries = pd.read_csv('data/allcountries.csv')
vac_countries = pd.read_csv('data/vacallcountries.csv')
```

```python
all_countries.head()
```

```python
vac_countries.head()
```

```python
vac_countries = vac_countries.rename(columns={'timeline.10/24/22': 'vaccinations'})
```

```python
vac_countries.head()
```

```python
result = pd.merge(all_countries, vac_countries, on="country")
```

```python
result.head()
```

```python
result['vaccination_percentage'] = (result['vaccinations'] / result['population']) * 100
```

```python
result.head()
```

```python
file_name = 'data/countries_processed.csv'

result.to_csv(file_name, index=False)
```
