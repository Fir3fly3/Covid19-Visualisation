```python
import requests
import pandas as pd
```

### Worldwide Stats

```python
# Convert an API JSON Response to a Python Dictionary

url = "https://disease.sh/v3/covid-19/all"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
data = response.json()
```

```python
worldwide = pd.DataFrame.from_dict(data, orient='index')
```

```python
worldwide = worldwide.T
```

```python
worldwide
```

```python
file_name = 'data/worldwide.csv'

worldwide.to_csv(file_name, index=False)
```

### All Countries

```python
url = 'https://disease.sh/v3/covid-19/countries'

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
data = response.json()
```

```python
# Convert a list of dictionaries using json_normalize().
allcountries=pd.json_normalize(data)
```

```python
allcountries.head()
```

```python
file_name = '//dallcountries.csv'

allcountries.to_csv(file_name, index=False)
```
